```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data_df = data.drop("Unnamed: 32", axis=1)
data_df.head()
```

```python
x = data_df.drop("diagnosis",axis=1)
y = data_df["diagnosis"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
model = KNeighborsClassifier(
    n_neighbors = 5,
    metric = "euclidean"
)
```

```python
model_x = KNeighborsClassifier(
    n_neighbors=5,
  metric = "manhattan"
  )
```

```python
accuracy = model_x.score(x_test,y_test)
accuracy1 = model.score(x_test,y_test)
```

```python
print(accuracy)
print(accuracy1)
```

```python
model_x.fit(x_train,y_train)
y_p = model_x.predict(x_test)
accuracy_model = accuracy_score(y_p,y_test)
print(accuracy_model)
```
